```python
from sklearn.manifold import TSNE
```

```python
 import pandas as pd
 import matplotlib.pyplot as plt
 import sklearn
 from sklearn import datasets
```

```python
sk_mnist=sklearn.datasets.load_digits()
```

```python
sk_data=sk_mnist.data
```

```python
model= TSNE(n_components=2, random_state=0)
#def def __init__(n_components=2, perplexity=30.0, 
  # early_exaggeration=12.0, learning_rate=200.0, 
  # n_iter(maximum)=1000, n_iter_without_progress=300, 
  # min_grad_norm=1e-07, metric='euclidean', init='random', 
  # verbose=0, random_state=None, method='barnes_hut', 
  # angle=0.5, n_jobs=None)

tsne_data = model.fit_transform(sk_data)
```

```python
tsne_data
```

```python
tsne_df = pd.DataFrame(data=tsne_data, columns=("PC1", "PC2"))

plt.scatter(tsne_df.PC1, tsne_df.PC2)
plt.title('t-SNE over MNIST')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

```python
ITERATIONS = 5000
PERPLEXITY = 50
```

```python
model= TSNE(n_components=2, random_state=0, n_iter=ITERATIONS, perplexity=PERPLEXITY)

tsne_data = model.fit_transform(sk_data)
```

```python
tsne_df = pd.DataFrame(data=tsne_data, columns=("PC1", "PC2"))

plt.scatter(tsne_df.PC1, tsne_df.PC2)
plt.title('t-SNE over MNIST')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

```python
model= TSNE(n_components=3, random_state=0, n_iter=ITERATIONS, perplexity=PERPLEXITY)

tsne_data = model.fit_transform(sk_data)

tsne_df = pd.DataFrame(data=tsne_data, columns=("PC1", "PC2","PC3"))

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(tsne_df.PC1, tsne_df.PC2, tsne_df.PC3)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

plt.show()
```
